```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
plt.figure(figsize=(20,10))
plt.subplot(3,3,1)
plt.plot(x,tan, "b", label="Grafik tans")
plt.legend(loc=9)
plt.subplot(3,3,2)
plt.plot(x,sin, "r", label="Grafik sinx")
plt.legend(loc=9)
plt.subplot(3,3,3)
plt.plot(x,cos, "g", label="Grafik cosx")
plt.legend(loc=9)
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
plt.plot(x,y)
plt.plot(x,y2)
plt.plot(x,y3)
plt.plot(x,y4)
plt.plot(x,y5)
```
